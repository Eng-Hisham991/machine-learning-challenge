# Import Dependencies

```python
import warnings
warnings.simplefilter('ignore')

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
a = df[(df['koi_disposition']=='CANDIDATE') | (df['koi_disposition']=='FALSE POSITIVE')]

a.head()
```

# Select features (columns)

```python
columns1= ['koi_disposition', 'koi_fpflag_nt', 'koi_fpflag_ss', 'koi_fpflag_co',
       'koi_fpflag_ec', 'koi_period', 
       'koi_time0bk', 'koi_impact',
       'koi_duration',
       'koi_depth', 'koi_prad', 
       'koi_teq', 'koi_insol', 
       'koi_model_snr', 'koi_tce_plnt_num', 'koi_steff', 
        'koi_slogg', 
       'koi_srad', 'ra', 'dec',
       'koi_kepmag']
df1 = a[columns1]
df1.head()
```

```python
#Use `koi_disposition` for the y values

X = df1.drop('koi_disposition', axis=1)
y = df1['koi_disposition']#.values.reshape(-1,1)
print(X.shape, y.shape)
```

# Create a Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42,stratify=y)
```

```python
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
model = SVC(kernel='linear')
```

```python
model.fit(X_train_scaled, y_train)
```

## Validate the model using the test data

```python
predictions = model.predict(X_test_scaled)
```

```python
print(f"Training Data Score: {model.score(X_train, y_train)}")
print(f"Testing Data Score: {model.score(X_test, y_test)}")
```

```python
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
```

```python
print(classification_report(y_test, predictions,
                target_names=['CANDIDATE', 'FALSE POSITIVE']))
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
from sklearn.model_selection import GridSearchCV
# param_grid = {'C': [ 200, 250, 300],
#               'gamma': [0.000000001,0.00000001,0.0000001]}
param_grid = {'kernel':('linear', 'rbf'), 'C':(1,0.25,0.5,0.75),'gamma': (1,2,3,'auto'),'decision_function_shape':('ovo','ovr'),'shrinking':(True,False)}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
# Make predictions with the hypertuned model
predictions = grid.predict(X_test_scaled)
print(classification_report(y_test, predictions,
                target_names=['CANDIDATE','FALSE POSITIVE']))
```
